Q1. Occurrence of substring in given string.

```
 
str1 = "what we think we become, we are python programmer"
print(len(str1))
```

```
print(str1.find("we"))
```

```
 print(str1.rfind("we"))
```

```
 str1 = "what we think we become, we are python programmer" 
sub = "we"
print(str1.index(sub))
```

```
 str1 = "what we think we become, we are python programmer"
print(str1.index("we", 0, -1))
```

```
 
str1 = "what we think we become, we are python programmer"
print(str1.count("we",0))
```

```
line = [i for i in range(len(str1)) if str1.startswith(sub,i)]
print(" The starting index's are", line)
```
